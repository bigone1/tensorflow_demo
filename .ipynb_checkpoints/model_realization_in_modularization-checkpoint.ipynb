# 网络的模块化实现
主要就是把生成数据、前向网络、反向传播，分别写成不同的文件，import来用，这里就省略了import，其实省略了import就和其他notebook写法差不太多。

```python
#文件名generates.py
import numpy as np
import matplotlib.pyplot as plt
seed = 2
def generates():
    rdm = np.random.RandomState(seed)
    X = rdm.randn(300,2)
    Y_ = [int(x0**2 + x1**2 < 2) for(x0,x1) in X]
    Y_c = [['red' if y else 'blue'] for y in Y_]
    X = np.vstack(X).reshape(-1,2)
    Y_ = np.vstack(Y_).reshape(-1,1)
    return X,Y_,Y_c
X, Y_, Y_c = generates()
print(X.shape)
print(Y_.shape)
print(len(Y_c))
```

```python
#文件名 net.py
import tensorflow as tf
def get_weight(shape, scale):#weights都纳入losses
    w = tf.Variable(tf.random_normal(shape), dtype = tf.float32)
    tf.add_to_collection('regular_losses', tf.contrib.layers.l2_regularizer(scale)(w))#避免歧义，改名regular_losses
    return w
def get_bias(shape):#b肯定是不能放在losses的，此losses非彼losses，这个losses是专门用来正则化的，正则化只对w，对b意义不大。
    b = tf.Variable(tf.constant(0.01, shape=shape))
    return b
def net(x, scale):
    w1 = get_weight([2,10], scale)
    b1 = get_bias([10])
    y1 = tf.nn.relu(tf.matmul(x,w1) + b1)
    w2 = get_weight([10,1],scale)
    b2 = get_bias([1])
    y = tf.matmul(y1,w2) + b2
    return y
```

```python
#文件名：train.py
# import net
# import generates
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

STEPS = 40000
BATCH_SIZE = 30
LEARNING_RATE_BASE = 0.001
LEARNING_RATE_DECAY = 0.999
REGULAR_SCALE = 0.01

def train():
    x = tf.placeholder(tf.float32, shape=(None,2))
    y_ = tf.placeholder(tf.float32,shape=(None,1))
    X,Y_,Y_c = generates()
    y = net(x,REGULAR_SCALE)
    global_step = tf.Variable(0,trainable = False)
    
    learning_rate = tf.train.exponential_decay(LEARNING_RATE_BASE,global_step,
                                              300 / BATCH_SIZE, LEARNING_RATE_DECAY,staircase=True)
    loss_mse = tf.reduce_mean(tf.square(y-y_))
    loss_total = loss_mse + tf.add_n(tf.get_collection('regular_losses'))
    
    train_op = tf.train.AdamOptimizer(learning_rate).minimize(loss_total)
    
    with tf.Session() as sess:
        tf.global_variables_initializer().run()
        for i in range(STEPS):
            start = (i*BATCH_SIZE) % 300
            end = start + BATCH_SIZE
            
            sess.run(train_op,feed_dict={x:X[start:end],y_:Y_[start:end]})
            if i % 2000 == 0:
                loss_v = sess.run(loss_total,feed_dict={x:X,y_:Y_})
                print('after %f steps, total loss is %f'%(i,loss_v))
        xx,yy = np.mgrid[-3:3:.01,-3:3:.01]
        grid = np.c_[xx.ravel(), yy.ravel()]
        probs = sess.run(y, feed_dict = {x:grid})
        probs = probs.reshape(xx.shape)
    plt.scatter(X[:,0],X[:,1],c=np.squeeze(Y_c))
    plt.contour(xx,yy,probs,levels=[.5])
    plt.show()
    
if __name__ == '__main__':
    train()
```
