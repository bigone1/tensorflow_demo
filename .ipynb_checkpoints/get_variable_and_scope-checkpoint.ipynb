```python
with tf.variable_scope('V1'):
	a1 = tf.get_variable(name='a1', shape=[1], initializer=tf.constant_initializer(1))
	a2 = tf.Variable(tf.random_normal(shape=[2,3], mean=0, stddev=1), name='a2')
with tf.variable_scope('V2'):
	a3 = tf.get_variable(name='a1', shape=[1], initializer=tf.constant_initializer(1))
	a4 = tf.Variable(tf.random_normal(shape=[2,3], mean=0, stddev=1), name='a2')
  
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables())
    writer = tf.summary.FileWriter('scope/variable_scope',tf.get_default_graph())
    print(a1.name)
    print(a2.name) 
    print(a3.name) 
    print(a4.name) 
#计算图，V1和V2下各有一个a1和a2
```

# 应用
必须变量名完全一致才是reuse，这个完全一致是包括scope的全名
想通过get_variable获取之前定义好的变量，必须用reuse=True
想通过get_variable新建，必须reuse=False
本例示意get_variable的用途，相当于引用，v,vs_v,vs_v1都指向同一个变量，打印一样，修改也会互相影响

```python
import tensorflow as tf;  
#获取变量的方式主要有以下两种，实践中tf.get_variable产生的变量一定要搭配tf.variable_scope使用，不然运行脚本会报错
v = tf.get_variable('v',shape= [1],initializer = tf.constant_initializer(1.0))
#使用直接定义变量不会报错，可以一直调用
#而且还没影响本来的v的值，所以这个是print(vc)    <tf.Variable 'v_1:0' shape=(1,) dtype=float32_ref>
vc = tf.Variable(tf.constant(3.0,shape = [1]),name = 'v')

#以下使用with语法，将tf.get_variable与tf.variable_scope搭配使用,且reuse=True时，之前必须定义V
#with tf.variable_scope('vs1',reuse = True):#vs1/v不是v，所以想重用之前的v，不能有名字
with tf.variable_scope('',reuse = True):
    vs_v = tf.get_variable('v',shape= [1],initializer = tf.constant_initializer(10.0))
    print(v)
    vs_v1 = tf.get_variable('v',shape= [1],initializer = tf.constant_initializer(20.0))#因为能get到，所以initializer是没用的
    #v12 = tf.get_variable('v12',shape= [1],initializer = tf.constant_initializer(2.0))##reuse为true的情况下，get_variable好像不能新建！！！！！！！！！！！！！！！
    print(vs_v1==vs_v)
    #计算图，v与v1下分别有变量v和v1
    writer = tf.summary.FileWriter('scope/variable_scope_reuse',tf.get_default_graph())
    
with tf.variable_scope('vs2',reuse = False):#vs2/v不是v，所以等于新建
    vs2_v = tf.get_variable('v',shape= [1],initializer = tf.constant_initializer(100.0))
    print(v)
#     v1 = tf.get_variable('v',shape= [1],initializer = tf.constant_initializer(2.0))#reuse=False，这时候就不能get v了
    vs2_v12 = tf.get_variable('v12',shape= [1],initializer = tf.constant_initializer(200.0))#因为不能get到，所以initializer启用


with tf.Session() as sess:
    tf.global_variables_initializer().run()
    print('no scope')
    print(sess.run(v))
    print(sess.run(vc))
    print('vs space')
    print(sess.run(vs_v))
    print(sess.run(vs_v1))
    print('vs2')
    print(sess.run(vs2_v))
    print(sess.run(vs2_v12))
    sess.run(tf.assign(v,[10.0]))
    print('after assign 10 to v')
    print(sess.run(v))
    print(sess.run(vc))
    print('vs none')
    print(sess.run(vs_v))
    print(sess.run(vs_v1))
    print('vs2')
    print(sess.run(vs2_v))
    print(sess.run(vs2_v12))

    sess.run(tf.assign(vs_v1,[11.0]))
    print('after assign 11 to vs_v1')
    print('no scope')
    print(sess.run(v))
    print(sess.run(vc))
    print('vs space')
    print(sess.run(vs_v))
    print(sess.run(vs_v1))
    print('vs2')
    print(sess.run(vs2_v))
    print(sess.run(vs2_v12))
    sess.run(tf.assign(v,[12.0]))
    print('after assign 12 to v')
    print(sess.run(v))
    print(sess.run(vc))
    print('vs none')
    print(sess.run(vs_v))
    print(sess.run(vs_v1))
    print('vs2')
    print(sess.run(vs2_v))
    print(sess.run(vs2_v12))
```

```python
import tensorflow as tf
with tf.variable_scope('foo'):
    v = tf.get_variable('v', [1], initializer = tf.constant_initializer(1.0))
# with tf.variable_scope('foo'):
#     v = tf.get_variable('v',[1])#already exist
with tf.variable_scope('foo', reuse=True):
    v = tf.get_variable('v', [1])
```

```python
#嵌套，其实对于名称来说本质没区别，不过就是名称多了一个斜杠。
#但是有一个上下文和继承关系。
import tensorflow as tf
with tf.variable_scope('root'):
    print(tf.get_variable_scope().reuse)#默认False
    print(tf.get_variable_scope().name)#打印一下name
    with tf.variable_scope('foo', reuse = True):
        print(tf.get_variable_scope().reuse)#手动设置的True
        print(tf.get_variable_scope().name)#打印一下name
        with tf.variable_scope('bar'):
            print(tf.get_variable_scope().reuse)#这个是继承的，上一层是True，现在也是True了，所以，root也可以看作继承了更外层的False。
            print(tf.get_variable_scope().name)#打印一下name
            
    #设置为True的上下文已经退出，恢复False
    print(tf.get_variable_scope().reuse)
    print(tf.get_variable_scope().name)#打印一下name
```

```python
#打印变量名
import tensorflow as tf
with tf.variable_scope('bar'):
    v0 = tf.get_variable('v', [1], initializer = tf.constant_initializer(1.0))
    print(v0.name)
with tf.variable_scope('bar',reuse=True):
    v2 = tf.get_variable('v',[1])
    print(v2.name)
```

## 应用：实现一个用get_variable的模型

```python
#基本写法
import tensorflow as tf
weights1 = tf.Variable(tf.truncated_normal([INPUT_NODE,LAYER1_NODE],stddev = 0.1))
biases1 = tf.Variable(tf.constant(0.1,shape=[LAYER1_NODE]))

weights2 = tf.Variable(tf.truncated_normal([LAYER1_NODE,OUTPUT_NODE],stddev = 0.1))
biases2 = tf.Variable(tf.constant(0.1,shape=[OUTPUT_NODE]))

def inference(input_tensor, avg_class, weights1, biases1, weights2, biases2):
    if avg_class == None:
        layer1 = tf.nn.relu(tf.matmul(input_tensor,weights1) + biases1)
        return tf.matmul(layer1,weights2) + biases2
    else:#可以看一下滑动平均中inference的写法
        #weights1为什么没用滑动平均，可能是忘了？
        layer1 = tf.nn.relu(tf.matmul(input_tensor, weights1) + avg_class.average(biases1))
        return tf.matmul(layer1,avg_class.average(weight2))+avg_class.average(biases2)
        
```

```python
#改进算法
#这个好处就是通过get_variable获得变量，不用把weights1,biases1,weights2,biases2通通当变量传入
#可读性强
import tensorflow as tf
import numpy as np
INPUT_NODE = 1
LAYER1_NODE = 10
OUTPUT_NODE = 1
with tf.variable_scope('layer1'):
    weights1 = tf.Variable(tf.truncated_normal([INPUT_NODE,LAYER1_NODE],stddev = 0.1))
    biases1 = tf.Variable(tf.constant(0.1,shape=[LAYER1_NODE]))

with tf.variable_scope('layer2'):
    weights2 = tf.Variable(tf.truncated_normal([LAYER1_NODE,OUTPUT_NODE],stddev = 0.1))
    biases2 = tf.Variable(tf.constant(0.1,shape=[OUTPUT_NODE]))

def inference(input_tensor,reuse = False):
    with tf.variable_scope('layer1',reuse = reuse):
        weights = tf.get_variable('weights',[INPUT_NODE,LAYER1_NODE],initializer=tf.truncated_normal_initializer(stddev=0.1))
        biases = tf.get_variable('biases',[LAYER1_NODE],initializer=tf.constant_initializer(0.0))
        layer1 = tf.nn.relu(tf.matmul(input_tensor,weights)+biases)
    with tf.variable_scope('layer2',reuse = reuse):
        weights = tf.get_variable('weights',[LAYER1_NODE,OUTPUT_NODE],initializer=tf.truncated_normal_initializer(stddev=0.1))
        biases = tf.get_variable('biases',[OUTPUT_NODE],initializer=tf.constant_initializer(0.0))
        layer2 = tf.nn.relu(tf.matmul(layer1,weights)+biases)
    return layer2
x = tf.placeholder(tf.float32,[None,INPUT_NODE],name='x-input')
y_ = tf.placeholder(tf.float32,[None,INPUT_NODE],name='y-input')
y = inference(x)

loss = tf.reduce_mean(tf.square(y_ - y))
train_op = tf.train.GradientDescentOptimizer(0.001).minimize(loss)
```

```python
#测试数据
train_x = np.random.standard_normal([300,1],) + 1#设置均值1,为了训练效果更好。
train_y = train_x * 2 + 3
print(np.mean(train_x))

# print(train_y)
```

```python
#训练并且使用训练好的参数来预测
with tf.Session() as sess:
    tf.global_variables_initializer().run()
    for i in range(1000):
        sess.run(train_op,feed_dict = {x:train_x, y_:train_y})
        if i % 100 == 0:
            loss_val = sess.run(loss,feed_dict = {x:train_x, y_:train_y})
            print('after %f steps, the loss is %f'%(i,loss_val))
    

    new_x = np.array([[1.],[3.]],dtype=np.float32)
    new_y = inference(new_x, True)
    print(new_x)
    print(new_y)
    print(sess.run(new_y))
```
